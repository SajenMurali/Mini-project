```python
import seaborn as sns
import matplotlib.pyplot as plt

# Data
models = [
    'RF (Sleep Quality)',
    'ANN (Sleep Quality)',
    'RF (Sleep Duration)',
    'ANN (Sleep Duration)'
]

accuracies = [
    0.48,
    0.54,
    0.63,
    0.53
]

# Create bar plot
plt.figure(figsize=(8,5))
sns.barplot(x=models, y=accuracies)

# Labels and title
plt.ylim(0, 1)
plt.ylabel("Mean CV Accuracy")
plt.xlabel("Models")
plt.title("Model Performance Comparison")
plt.xticks(rotation=20)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Confusion matrix values from previous notebook
cm = np.array([
    [4, 5],
    [6, 5]
])

labels = ["Poor Sleep", "Good Sleep"]
```

```python
plt.figure(figsize=(5,4))
plt.imshow(cm, cmap="Blues")

# Add numbers inside boxes
for i in range(2):
    for j in range(2):
        plt.text(j, i, cm[i, j],
                 ha="center", va="center", fontsize=12)

plt.xticks([0, 1], labels)
plt.yticks([0, 1], labels)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Random Forest  Sleep Quality")

plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
cm_ann = np.array([
    [0, 0, 5],
    [0, 0, 4],
    [1, 0, 10]
])

labels = ["Class 0", "Class 1", "Class 2"]
```

```python
plt.figure(figsize=(5,4))
plt.imshow(cm_ann, cmap="Greens")

# Add numbers inside cells
for i in range(cm_ann.shape[0]):
    for j in range(cm_ann.shape[1]):
        plt.text(j, i, cm_ann[i, j],
                 ha="center", va="center", fontsize=12)

plt.xticks(range(3), labels)
plt.yticks(range(3), labels)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("ANN  Sleep Duration")

plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
cm = np.array([
    [1, 1, 3],
    [0, 0, 4],
    [2, 3, 6]
])

# Class labels
labels = ["Class 0", "Class 1", "Class 2"]

# Plot confusion matrix
plt.figure(figsize=(5, 4))
plt.imshow(cm, cmap="Blues")

# Add values inside each cell
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j],
                 ha="center", va="center", fontsize=12)

# Axis labels and ticks
plt.xticks(range(len(labels)), labels)
plt.yticks(range(len(labels)), labels)
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Random Forest  Sleep Duration")

# Color bar
plt.colorbar()
plt.tight_layout()
plt.show()
```
