```python
import pandas as pd
import numpy as np
```

```python
# load dataset
df = pd.read_csv("data - Sheet1.csv")

# create target variable
df['Target'] = df['SleepQuality'].apply(lambda x: 0 if x <= 3 else 1)

# view result
df[['SleepQuality', 'Target']].head()
```

```python
df.head()
```

```python
# rename column for clarity (recommended)
df.rename(columns={'sleep duration hours ': 'SleepDuration_min'}, inplace=True)
```

```python
# create categorical target based on sleep duration
def sleep_duration_category(x):
    if x >= 420:
        return 2   # Good sleep
    elif x >= 360:
        return 1   # Moderate sleep
    else:
        return 0   # Poor sleep

df['SleepDuration_Target'] = df['SleepDuration_min'].apply(sleep_duration_category)

# view result
df[['SleepDuration_min', 'SleepDuration_Target']].head()
```

```python
df.head()
```

```python
# save modified dataset
df.to_csv("sleep_dataset_modified.csv", index=False)
```

```python
data=pd.read_csv("sleep_dataset_modified.csv")
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# create screentime in minutes
data['screentime_min'] = data['ScreenTime (Total Hours)'] * 60

# optional: round
data['screentime_min'] = data['screentime_min'].round(0)

data[['ScreenTime (Total Hours)', 'screentime_min']].head()
```

```python
data.info()
```

```python
data.shape
```

```python
data.head()
```

```python
print("Initial shape:", data.shape)

# -------------------------------
# 2. RENAME COLUMNS (CLEAN NAMES)
# -------------------------------
data.columns = data.columns.str.strip().str.replace(" ", "_").str.lower()

# Explicit renaming for clarity
data.rename(columns={
    'bed(minutes)': 'bedtime_min',
    'wakeup(adjusted)': 'wakeup_min'
}, inplace=True)

# -------------------------------
# 3. HANDLE MISSING VALUES
# -------------------------------
num_cols = data.select_dtypes(include=['int64', 'float64']).columns
data[num_cols] = data[num_cols].fillna(data[num_cols].median())

# -------------------------------
# 4. REMOVE DUPLICATES
# -------------------------------
data.drop_duplicates(inplace=True)

# -------------------------------
# 5. REMOVE INVALID SLEEP RECORDS
# -------------------------------
data = data[(data['sleepduration_min'] >= 120) & (data['sleepduration_min'] <= 720)]

# -------------------------------
# 6. CREATE TARGET VARIABLE
# -------------------------------
if 'target' not in data.columns:
    data['target'] = data['sleepquality'].apply(lambda x: 0 if x <= 3 else 1)

# -------------------------------
# 9. FEATURE SCALING
# -------------------------------
features_to_scale = [
    'screentime_min',
    'socialmediausage_(minutes)',
    'unlocks',
    'sleepduration_min'
]

scaler = StandardScaler()
data[features_to_scale] = scaler.fit_transform(data[features_to_scale])

# -------------------------------
# 10. FINAL CHECK
# -------------------------------
print("Final shape:", data.shape)
print(data.head())
print(data.info())

# -------------------------------
# 11. SAVE FILE
# -------------------------------
data.to_csv("sleep_dataset_cleaned_preprocessed.csv", index=False)

print("✅ Dataset cleaned and saved successfully!")
```

```python
# FEATURES & TARGET
X1 = data[
    [
        'screentime_min',
        'socialmediausage_(minutes)',
        'unlocks'
    ]
]

y1 = data['target']

# keep only numeric columns
X1 = X1.select_dtypes(include=['int64', 'float64'])

# split
from sklearn.model_selection import train_test_split
X1_train, X1_test, y1_train, y1_test = train_test_split(
    X1, y1, test_size=0.2, random_state=42, stratify=y1
)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

rf1 = RandomForestClassifier(n_estimators=100, random_state=42)
rf1.fit(X1_train, y1_train)

y1_pred = rf1.predict(X1_test)

print("Accuracy:", accuracy_score(y1_test, y1_pred))
print("Confusion Matrix:\n", confusion_matrix(y1_test, y1_pred))
```

```python
# Clean feature set (no leakage)
X_ann = data[['screentime_min', 'socialmediausage_(minutes)', 'unlocks']]
y_ann = data['target']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_ann, y_ann,
    test_size=0.2,
    random_state=42,
    stratify=y_ann
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

ann = Sequential([
    Dense(16, activation='relu', input_shape=(X_train.shape[1],)),
    Dense(8, activation='relu'),
    Dense(1, activation='sigmoid')
])

ann.compile(
    optimizer=Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

history = ann.fit(
    X_train, y_train,
    epochs=50,
    batch_size=8,
    validation_split=0.2,
    verbose=1
)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred_ann = (ann.predict(X_test) > 0.5).astype(int).flatten()

print("ANN Accuracy:", accuracy_score(y_test, y_pred_ann))
print("ANN Confusion Matrix:\n", confusion_matrix(y_test, y_pred_ann))
print("ANN Classification Report:\n",
      classification_report(y_test, y_pred_ann))
```

```python
from sklearn.model_selection import KFold
import numpy as np

kf = KFold(n_splits=5, shuffle=True, random_state=42)
ann_cv_acc = []

for train_idx, test_idx in kf.split(X_ann):
    X_tr, X_te = X_ann.iloc[train_idx], X_ann.iloc[test_idx]
    y_tr, y_te = y_ann.iloc[train_idx], y_ann.iloc[test_idx]

    model = Sequential([
        Dense(16, activation='relu', input_shape=(X_tr.shape[1],)),
        Dense(8, activation='relu'),
        Dense(1, activation='sigmoid')
    ])

    model.compile(
        optimizer=Adam(0.001),
        loss='binary_crossentropy',
        metrics=['accuracy']
    )

    model.fit(X_tr, y_tr, epochs=40, batch_size=8, verbose=0)
    y_pred = (model.predict(X_te) > 0.5).astype(int).flatten()
    ann_cv_acc.append(accuracy_score(y_te, y_pred))

print("ANN CV Accuracies:", ann_cv_acc)
print("Mean ANN CV Accuracy:", np.mean(ann_cv_acc))
```

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True, random_state=42)
ann_scores = []

for tr, te in kf.split(X_sd):
    X_tr, X_te = X_sd.iloc[tr], X_sd.iloc[te]
    y_tr, y_te = y_sd.iloc[tr], y_sd.iloc[te]

    model = Sequential([
        Dense(32, activation='relu', input_shape=(X_tr.shape[1],)),
        Dense(16, activation='relu'),
        Dense(3, activation='softmax')
    ])

    model.compile(
        optimizer=Adam(0.001),
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    model.fit(X_tr, y_tr, epochs=40, batch_size=8, verbose=0)
    preds = model.predict(X_te).argmax(axis=1)
    ann_scores.append(accuracy_score(y_te, preds))

print("ANN CV Accuracies:", ann_scores)
print("Mean ANN CV Accuracy:", np.mean(ann_scores))
```
